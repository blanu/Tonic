```python
import math
import wave
import numpy as np
from IPython.display import display

# FSK type modem
# Goals:

def fft_bins_for_frequency(target_frequency, sample_rate):
    """
    Calculate the number of FFT bins required to detect a signal of a given frequency.

    Parameters:
        target_frequency (float): The frequency of the signal in Hz.
        sample_rate (float): The sampling rate in Hz.

    Returns:
        int: Number of FFT bins required.
    """
    if target_frequency <= 0 or sample_rate <= 0:
        raise ValueError("Target frequency and sample rate must be positive values.")

    # Calculate the minimum time period required to detect the target frequency
    min_time = 1 / target_frequency  # Time period of one cycle of the signal

    # Calculate the minimum number of samples required to observe this time period
    min_samples = int(np.ceil(min_time * sample_rate))

    # The number of FFT bins is equivalent to the number of samples
    fft_bins = min_samples

    return fft_bins

def differentiate_frequencies(freq1, freq2, sample_rate):
    """
    Calculate the minimum number of samples and FFT bins required to differentiate two frequencies.

    Parameters:
        freq1 (float): First frequency in Hz.
        freq2 (float): Second frequency in Hz.
        sample_rate (float): Sampling rate in Hz.

    Returns:
        tuple: (minimum_samples, fft_bins)
    """
    # Ensure frequencies are positive
    freq1, freq2 = abs(freq1), abs(freq2)

    # Calculate the frequency resolution required to differentiate the signals
    delta_freq = abs(freq1 - freq2)
    if delta_freq == 0:
        raise ValueError("The two frequencies must be different to differentiate them.")

    # Calculate the minimum observation time (T) needed to distinguish the frequencies
    min_time = 1 / delta_freq

    # Minimum number of samples is the observation time multiplied by the sampling rate
    min_samples = int(np.ceil(sample_rate * min_time))

    return min_samples






### modem settings

# output sample rate for wav file or to send to DAC
outSampleRate = 44100 # Hz
output_file = 'tonic-z.wav'
volume = 0.5 # 50%

# bits per second
targetGoodPut = 50

# Error Correction Factor, i.e. bits added for FEC, overhead, etc
errorCorrectionFactor = 2

# slowdown factor, to help visualize tones and/or increase performance
slowdownFactor = 1

# how many bits are represented by a tone
# the message must be evenly divisible by this value or the message will need padding.
bitsPerTone = 2

# number of tones to use
toneCount = 2 ** bitsPerTone

# symbol rate also known as baud rate, how many tones per second do we generate
symbolRate = (targetGoodPut  * errorCorrectionFactor) / bitsPerTone

# enable spreading
slowSpreading = False
spreadingBandwidth = 10000 # Hz
minHopDistance = 0 # Hz, calculated further down if used

# tone spacing in Hz, for Orthogonal FSK min spacing = 1/2 *T for phase
# orthogonal requires n/T spacing where n is an integer>0
# T = baud rate, symbolRate
# if phase coherent then toneSpacing = 1/2 * symbolRate; otherwise toneSpacing = symbolRate
# Ref Wireless Digital Comm, pg 33, 43, 44
toneSpacing = 1/2 * symbolRate
totalBandwidth = toneSpacing * toneCount

# Valid frequency range in Hz that tones need to fall in to be sent over a channel
basebandMin = 300
basebandMax = 2200

# Hz, shift the tones up to be in the passband of a common radio
# basebandCarrierFrequency = n * 1/2 * toneSpacing # where n is an integer>0
# Ref Wireless Digital Comm, pg 46
basebandCarrierFrequency = 1000
minTone = basebandCarrierFrequency - (totalBandwidth/2) + toneSpacing/2
maxTone = basebandCarrierFrequency + (totalBandwidth/2) - toneSpacing/2

# make a list of valid tones
tones = []
for i in range(toneCount):
    tones.append((minTone + i * toneSpacing))

# apply slowdown factor
symbolRate = symbolRate * slowdownFactor

# if slowSpreading:
#     # these are not correct, revisit when implementing spreading
#     basebandCarrierFrequency = spreadingBandwidth/2 + toneSpacing * 2
#     minTone = basebandCarrierFrequency - totalBandwidth/2
#     maxTone = basebandCarrierFrequency + totalBandwidth/2
#     minHopDistance = toneSpacing + totalBandwidth
#     maxHopDistance = spreadingBandwidth - minHopDistance

# Print Modem Config Variables
print("targetGoodPut: ", targetGoodPut, "bps")
print("errorCorrectionFactor: ", errorCorrectionFactor)
print("bitsPerTone: ", bitsPerTone)
print("toneCount: ", toneCount)
print("symbolRate: ", symbolRate, "Baud")
print("slowSpreading: ", slowSpreading)
print("toneSpacing: ", toneSpacing, "Hz")
print("totalBandwidth: ", totalBandwidth, "Hz")
print("basebandCarrierFrequency: ", basebandCarrierFrequency, "Hz")
print("minTone: ", minTone, "Hz")
print("maxTone: ", maxTone, "Hz")
print("Tones Used:", tones)





def combine_bits_to_decimal(bits, chunk_size):
    """
    Combine bits from a list into decimal values.

    :param bits: List of 0s and 1s.
    :param chunk_size: Number of bits to combine for each decimal value.
    :return: List of integers representing the decimal values.
    """
    if chunk_size <= 0:
        raise ValueError("Chunk size must be a positive integer.")

    # Ensure the length of the bits list is a multiple of chunk_size
    if len(bits) % chunk_size != 0:
        raise ValueError("The length of the bits list must be divisible by chunk_size.")

    decimal_values = []
    for i in range(0, len(bits), chunk_size):
        chunk = bits[i:i+chunk_size]  # Extract a chunk of bits
        decimal_value = int(''.join(map(str, chunk)), 2)  # Convert chunk to decimal
        decimal_values.append(decimal_value)

    return decimal_values


def generate_sine_wave(frequency, sample_rate, duration):
    """
    Generate a sine wave as a numpy array with the duration adjusted to whole cycles.

    Args:
        frequency (float): Frequency of the sine wave in Hz.
        sample_rate (int): Samples per second.
        duration (float): Desired duration of the sine wave in seconds.

    Returns:
        numpy.ndarray: Array containing the sine wave samples.
    """
    # Calculate the number of cycles in the given duration
    total_cycles = duration * frequency

    # Round to the nearest whole number of cycles
    rounded_cycles = round(total_cycles)

    # Calculate the adjusted duration based on the rounded cycles
    adjusted_duration = rounded_cycles / frequency

    # Calculate the total number of samples for the adjusted duration
    total_samples = int(adjusted_duration * sample_rate)

    # Generate the time vector for the sine wave
    t = np.linspace(0, adjusted_duration, total_samples, endpoint=False)

    # Generate the sine wave
    sine_wave = np.sin(2 * np.pi * frequency * t)

    return sine_wave

def save_wave_file(filename, wave_data, sample_rate):
    """
    Save a numpy array as a wave file.

    Args:
        filename (str): Output filename.
        wave_data (numpy.ndarray): Waveform data to save.
        sample_rate (int): Sampling rate in Hz (default is 44100 Hz).
    """

    # Write to a wave file
    with wave.open(filename, 'w') as wav_file:
        wav_file.setnchannels(1)  # Mono
        wav_file.setsampwidth(2)  # 16-bit
        wav_file.setframerate(sample_rate)
        for data in wave_data:
            # Convert waveform data to 16-bit PCM format
            converted = (data * 32767).astype(np.int16)
            wav_file.writeframes(converted.tobytes())

def generate_random_bits(length):
    """
    Generate an array of random bits (0 or 1).

    Args:
        length (int): The length of the array.

    Returns:
        numpy.ndarray: Array of random bits.
    """
    return np.random.randint(0, 2, length, dtype=np.uint8)




### Example Usage

message = bitsFixedRandom
# for bit in message:
#     print(bit,end=",")
# print("")
chunkedMessage = combine_bits_to_decimal(message, bitsPerTone)
print("chunkedMessage: ", chunkedMessage)  # Output: [11, 12] (binary 1011 -> 11, 1100 -> 12)

tonesToTransmit = []

# assemble the list of tones used to construct the transmission
for value in chunkedMessage:
    tonesToTransmit.append(tones[value])
print("tonesToTransmit: ", tonesToTransmit)



symbol_duration = 1.0 / symbolRate  # Duration for one symbol

# Precompute all tones once
results = [
    generate_sine_wave(tone, outSampleRate, symbol_duration) * volume
    for tone in tonesToTransmit
]


# Save the sine wave as a wave file
save_wave_file(output_file, results, outSampleRate)
print(f"Generated audio file: {output_file}")

```
